```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class KalmanFilter(object):
    def __init__(self, dt, u, std_acc, std_meas):
        self.dt = dt
        self.u = u
        self.std_acc = std_acc
        #state transition model
        self.A = np.array([[1, self.dt],
                            [0, 1]])
        self.B = np.array([[(self.dt**2)/2], [self.dt]]) 
        self.H = np.array([[1,0]])
        self.Q = np.array([[(self.dt**4)/4, (self.dt**3)/2],
                            [(self.dt**3)/2, self.dt**2]]) * self.std_acc**2
        self.R = std_meas**2
        self.P = np.eye(2)
        self.x = np.array([[908],[u]])

    def predict(self):
          
          # Update time state
          self.x = np.dot(self.A, self.x) + np.dot(self.B, self.u)
          # Calculate error covariance
          # P= A*P*A' + Q
          self.P = np.dot(np.dot(self.A, self.P), self.A.T) + self.Q
          return self.x

    def update(self, z):
      
          # S = H*P*H'+R
          S = np.dot(self.H, np.dot(self.P, self.H.T)) + self.R
          # Calculate the Kalman Gain
          # K = P * H'* inv(H*P*H'+R)
          K = np.dot(np.dot(self.P, self.H.T), np.linalg.inv(S))  
          self.x = np.round(self.x + np.dot(K, (z - np.dot(self.H, self.x))))  
          I = np.eye(self.H.shape[1])
          self.P = (I - (K * self.H)) * self.P  

```

```python
import cv2 as cv
import time

Conf_threshold = 0.4
NMS_threshold = 0.4
COLORS = [(0, 255, 0), (0, 0, 255), (255, 0, 0),
          (255, 255, 0), (255, 0, 255), (0, 255, 255)]

class_name = []
with open('cfg/coco.names', 'r') as f:
    class_name = [cname.strip() for cname in f.readlines()]

#Creating the Detection Network using weights and configurations from Yolov4 -Tiny
net = cv.dnn.readNet('yolov4-tiny.weights', 'cfg/yolov4-tiny.cfg')
net.setPreferableBackend(cv.dnn.DNN_BACKEND_CUDA)
net.setPreferableTarget(cv.dnn.DNN_TARGET_CUDA)

#setting parameters using DetectionModel
model = cv.dnn_DetectionModel(net)
model.setInputParams(size=(960, 540), scale=1/255, swapRB=True)

#creating object for video capture 
cap = cv.VideoCapture('ball.mp4')

#for calculating the frames
starting_time = time.time()
frame_counter = 0
center= np.array([0, 276])

if cap.isOpened():
  width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
  height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
  res=(int(width), int(height))
  fourcc = cv.VideoWriter_fourcc(*'H264') #codec
  out = cv.VideoWriter('output.mp4', fourcc, 20.0, res)


  #creating the KalMan Object KF
  kf= KalmanFilter(0.006, 1, 1, 0.001)

  while True:
      #reading the video file 
      start_time=time.time()
      ret, frame = cap.read()
      frame_counter += 1
      if ret == False:
        break
      
      classes, score, boxes = model.detect(frame, Conf_threshold, NMS_threshold)
      
      if len(boxes)>0:

        for box in boxes :
            #creating bounding box for the detected circle in a frame
            color = COLORS[0]
            cv.rectangle(frame, box, color, 2)
            cv.putText(frame,'Ball', (box[0], box[1]-10), cv.FONT_HERSHEY_COMPLEX, 0.5, color, 1)

            # calculate x,y coordinate of center for which bounding box exist ( in ndarray form)
            center= np.array([box[0] + (box[2]) // 2, box[1] + ((box[3])) // 2])
            cv.circle(frame,(center), 10, (255, 255, 255), 2)   #white color to represent ground truth value
            
            #predict
            x = kf.predict()
            cv.circle(frame,(int(x[0]),center[1]), 10, (0, 0, 255), 2) #red color to indicate the predicted center

            #update
            z = kf.H * center[0]  + np.random.normal(0, 2)   #[[x], [200]]
            kf.update(z.item(0))
      else: 
        x = kf.predict()
        cv.circle(frame,(int(x[0]),center[1]), 10, (0, 0, 255), 2)
  
      out.write(frame)
  out.release()
end_time=time.time()
cap.release()
```

```python
#Multi Object Detection


def assign_object(center, tracker):
    #assigning the first tracker
    if len(tracker)==0:
      return 0
    
    #calculating the distance between the new center and the existing center
    dist=[]
    for i,j in tracker.items():
      dist.append(np.sqrt(np.sum(np.square(center - j))))

    #putting a threshold value to detect a new tracker, else returning the index of the closest object
    if min(dist)<420:
      return (dist.index(min(dist)))
    else:
      return len(tracker)

#setting parameters using DetectionModel
model = cv.dnn_DetectionModel(net)
model.setInputParams(size=(416, 416), scale=1/255, swapRB=True)

#creating object for video capture 
capture = cv.VideoCapture('ball_multi.mp4')

#for calculating the frames
starting_time_multi = time.time()
frame_counter_multi = 0
center= np.array([0, 276])

if capture.isOpened():
  width = int(capture.get(cv.CAP_PROP_FRAME_WIDTH))
  height = int(capture.get(cv.CAP_PROP_FRAME_HEIGHT))
  res=(int(width), int(height))
  fourcc = cv.VideoWriter_fourcc(*'H264') #codec
  mout = cv.VideoWriter('output_multi.mp4',fourcc, 20.0, res)

  tracker={}
  kf_tracker={}
  
  while True:
      #reading the video file 
      start_time=time.time()
      ret, frame = capture.read()
      frame_counter_multi += 1

      object_detected = list()

      if ret == False:
        break
      classes, score, boxes = model.detect(frame, Conf_threshold, NMS_threshold)

      if len(boxes)>0:

        for box in boxes :

            #calculate x,y coordinate of center for which bounding box exist ( in ndarray form) and get the tracker id for this box
            center= np.array([box[0] + (box[2]) // 2, box[1] + ((box[3])) // 2])
            obj_id= assign_object(center, tracker)
            object_detected.append(obj_id)

            #assigning a new Kalman filter if it is a new tracker, else updating the center for the previous tracker
            if obj_id >1:
              continue
            if obj_id not in tracker:
              tracker[obj_id]= center
              kf_tracker[obj_id]= KalmanFilter(0.006, 1, 1, 0.001)
            else:
              tracker[obj_id]= center
            

            #creating bounding box for the detected circle in a frame
            color = COLORS[obj_id]
            cv.rectangle(frame, box, color, 2)
            cv.putText(frame,'Object'+ str(obj_id+1), (box[0], box[1]-10), cv.FONT_HERSHEY_COMPLEX, 0.5, color, 1)
            cv.circle(frame,(center), 10, (255, 255, 255), 2)
            
            #predict
            kf= kf_tracker[obj_id]
            x = kf.predict()    #getting the relevant Kalman Object
            cv.circle(frame,(int(x[0]),center[1]), 10, (0, 0, 255), 2)
            
            #update
            z = kf.H * center[0]  + np.random.normal(0, 1)   
            kf.update(z.item(0))
      
      
      # else:  #incase the object isn't detected, we will update the value using Kalman Update Function
      #   for i,j in tracker.items():
      #     if i not in object_detected:
      #       x = kf_tracker[i].predict()
      #       cv.circle(frame,(int(x[0]),center[1]), 10, (0, 0, 255), 2)
      #       tracker[i]=x


      mout.write(frame)

  mout.release()
end_time=time.time()
capture.release()
```
